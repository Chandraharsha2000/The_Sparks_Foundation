Prediction using Decision Tree Algorithm

● Create the Decision Tree classifier and visualize it graphically.

● The purpose is if we feed any new data to this classifier, it would be able to predict the right class accordingly.

● Dataset : https://bit.ly/3kXTdox

Importing Libraries

```python
import pandas as pd # working with dataframe and importing dataset

from sklearn.tree import DecisionTreeClassifier # used to import decision tree classifier

from sklearn.model_selection import train_test_split #used to train and test the data 

from sklearn import metrics # used for getting the accuarcy score
```

Reading the Data

```python
col_names = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'Outcome']
df = pd.read_csv("Iris.csv", header=None, names=col_names)
print("Dataset Imported ")
```

Visualizing the Data

```python
df.info()
df = df.iloc[1:]
print()
print("The first 5 Tuples of the Data:")
df.head()
```

Preparing for Training and Testing Data

```python
feature_cols = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
X = df[feature_cols] 
y = df.Outcome
```

Splitting the Data

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

Accuracy Prediction using Decision Tree Classifer

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split 
from sklearn import metrics 
clf = DecisionTreeClassifier()
clf = clf.fit(X_train,y_train)
y_pred = clf.predict(X_test)
```

```python
print("Accuracy Obtained is:",(metrics.accuracy_score(y_test, y_pred)))
```

Graphical Representation of the data using Decison Tree Classifier

```python
import pydotplus
```

```python
from sklearn.tree import export_graphviz
from six import StringIO  
from IPython.display import Image  
from pydotplus import graph_from_dot_data

dot_data = StringIO()
export_graphviz(clf, out_file=dot_data,  
                filled=True, rounded=True,
                special_characters=True,feature_names = feature_cols,class_names=['0','1','2'])
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())
graph.write_png('Iris.png')
Image(graph.create_png())
```
